```python
#Filter data for chosen school and exam
df[['School Name', 'Regents Exam', 'Mean Score']]
```

```python
#Filter data for chosen school Icahn Charter School
icahn_data = df[(df['School Name'] == 'Icahn Charter School') & (df['Regents Exam'] == 'Common Core Algebra')]
```

```python
#Reassign 's' values
df['Mean Score'] = pd.to_numeric(df['Mean Score'], errors = 'coerce')
```

```python
#Drop all NaN values from mean score column
df.dropna(subset=['Mean Score'])
```

```python
#Display basic stats about columns for chosen data
icahn_data.describe()
```

```python
#Show max mean score for Common Core Algebra exam at Icahn Charter
icahn_data.groupby(['Regents Exam'])['Mean Score'].max()
```

```python
#Show min mean score for Common Core Algebra exam at Icahn Charter
icahn_data.groupby(['Regents Exam'])['Mean Score'].min()
```

```python
#Show average mean score for Common Core Algebra exam at Icahn Charter
icahn_data.groupby(['Regents Exam'])['Mean Score'].mean()
```

```python
#Compare average mean score for Common Core Algebra exam at all other schools
df.groupby(['Regents Exam'])['Mean Score'].mean()
```
